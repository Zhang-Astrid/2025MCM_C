```python
import pandas as pd
import numpy as np
```

```python
# 定义相对路径
file_data_dictionary = 'data/data_dictionary.csv'
file_summerOly_athletes = 'data/summerOly_athletes.csv'
file_summerOly_hosts = 'data/summerOly_hosts.csv'
file_summerOly_medal_counts = 'data/summerOly_medal_counts.csv'
file_summerOly_programs = 'data/summerOly_programs.csv'

# 读取表格
df_data_dictionary = pd.read_csv(
    file_data_dictionary,
    encoding='utf-8',
    encoding_errors='ignore'  # 忽略解码错误
)

df_summerOly_athletes = pd.read_csv(
    file_summerOly_athletes,
    encoding='utf-8',
    encoding_errors='ignore'  # 忽略解码错误
)
# 筛选并修改 'NOC' 列中 LBA 为 LBI
df_summerOly_athletes.loc[df_summerOly_athletes['NOC'] == 'LBA', 'NOC'] = 'LBI'

# 筛选并修改 'NOC' 列中 ROC 为 RUS
df_summerOly_athletes.loc[df_summerOly_athletes['NOC'] == 'ROC', 'NOC'] = 'RUS'


df_summerOly_hosts = pd.read_csv(
    file_summerOly_hosts,
    encoding='utf-8',
    encoding_errors='ignore'  # 忽略解码错误
)

df_summerOly_medal_counts = pd.read_csv(
    file_summerOly_medal_counts,
    encoding='utf-8',
    encoding_errors='ignore'  # 忽略解码错误
)

df_summerOly_programs = pd.read_csv(
    file_summerOly_programs,
    encoding='utf-8',
    encoding_errors='ignore'  # 忽略解码错误
)
```

```python
# 筛选 NOC 列，并去重
# 从 athletes 表中筛选 NOC
noc_athletes = df_summerOly_athletes[['NOC']].drop_duplicates()

# 从 medal_counts 表中筛选 NOC
noc_medal_counts = df_summerOly_medal_counts[['NOC']].drop_duplicates()

# 合并 NOC 并去重
all_nocs = pd.concat([noc_athletes, noc_medal_counts]).drop_duplicates().reset_index(drop=True)

# 输出结果
print("筛选出的唯一 NOC 值：")
all_nocs
```

```python
# 清理 Host 列中的国家名称
df_summerOly_hosts['Country_Name'] = df_summerOly_hosts['Host'].str.split(',').str[-1].str.strip()
host_Country = df_summerOly_hosts['Country_Name']
host_Country
```

```python
# 创建 Team 与 NOC 的对应关系
team_to_noc = df_summerOly_athletes[['Team', 'NOC']].drop_duplicates()
team_to_noc
```

```python
# 创建 NOC 与 Country_Name 的对应关系
noc_to_country = df_summerOly_medal_counts[['NOC']].drop_duplicates()
noc_to_country = noc_to_country.merge(team_to_noc, left_on='NOC', right_on='Team', how='left')
noc_to_country
```

```python
team_to_noc1 = df_summerOly_athletes[['NOC']].drop_duplicates()
team_to_noc1
```

```python
team_to_noc2 = pd.merge(team_to_noc1, noc_to_country, how='left', left_on='NOC', right_on='NOC_y',
                        suffixes=('_1', '_2'), right_index=False)
result = team_to_noc2[['NOC', 'NOC_x']].rename(columns={'NOC_x': 'Country_Name'})
result = pd.merge(result, host_Country, how='left', on='Country_Name', right_index=False)
result
```

```python
# 筛选出 Country_Name 为空的行
missing_country = result[result['Country_Name'].isnull()]

# 用 NOC 关联 df_summerOly_athletes 补全 Country_Name
missing_country = missing_country.merge(df_summerOly_athletes[['NOC', 'Team']], on='NOC', how='outer')
missing_country = missing_country.reset_index(drop=True).drop_duplicates()
missing_country
```

```python
# 用补全的 Team 列更新到原始 result 表
result1 = pd.merge(result, missing_country, how='outer', on='NOC',left_index=False, right_index=False)
result1['Country_Name_x'] = result1['Country_Name_x'].fillna(result1['Team'])
result1 = result1[['NOC', 'Country_Name_x']].rename(columns={'Country_Name_x': 'Country_Name'}).drop_duplicates()
result1
```

```python
# 按 NOC 分组并统计数量
grouped_counts = result1.groupby('NOC').size().reset_index(name='Count')

# 筛选数量大于 1 的分组
filtered_noc = grouped_counts[grouped_counts['Count'] > 1]

# 合并筛选后的分组结果回原表，获取所有行详细信息
result_filtered = result1[result1['NOC'].isin(filtered_noc['NOC'])]

# result_filtered.to_csv('data/duplicateCountryName.csv', index=False)
result_filtered
# 手动删除重复国家行
```

```python
# 筛选数量小于等于 1 的分组
valid_noc = grouped_counts[grouped_counts['Count'] <= 1]

# 保留原表中属于 valid_noc 的行
result1_filtered = result1[result1['NOC'].isin(valid_noc['NOC'])]
# 查看结果
result1_filtered
```

```python
df_washedCountry = pd.read_csv(
    'data/duplicateCountryName.csv',
    encoding='utf-8',
    encoding_errors='ignore'  # 忽略解码错误
)
df_washedCountry
```

```python
result1_appended = pd.concat([result1_filtered, df_washedCountry], ignore_index=True)
result1_appended
```

```python
df_summerOly_hosts
# 按国家分组，将主办年份汇总为列表
host_years = df_summerOly_hosts.groupby('Country_Name')['Year'].apply(list).reset_index()
host_years = host_years[['Year','Country_Name']]
# 替换特殊年份的 Country_Name
host_years.loc[host_years['Year'].apply(lambda x: 1916 in x), 'Country_Name'] = 'Germany'
host_years.loc[host_years['Year'].apply(lambda x: 1944 in x), 'Country_Name'] = 'United Kingdom'
host_years.loc[host_years['Year'].apply(lambda x: 1940 in x), 'Country_Name'] = 'Japan'
host_years.loc[host_years['Country_Name'].apply(lambda x: 'Japan (postponed to 2021 due to the coronavirus pandemic)' in x), 'Country_Name'] = 'Japan'
host_years.loc[host_years['Country_Name'].apply(lambda x: 'Germany' in x), 'Year'] = '[1916, 1936]'
host_years = host_years.drop_duplicates(subset=['Country_Name'], keep='first')
host_years
```

```python
# 使用 merge 方法合并两个表
countries = pd.merge(
    result1_appended,  # 包含 NOC 和 Country_Name 的主表
    host_years,        # 包含 Country_Name 和 Host_Years 的表
    on='Country_Name', # 通过 Country_Name 进行匹配
    how='left'         # 左连接，保留所有 NOC 的数据
)

# 确保最终表格包含的字段为 NOC、Country_Name、Host_Years
countries = countries[['NOC', 'Country_Name', 'Year']].rename(columns={'Year': 'Host_Years'})

# 查看最终构建的 Countries 表
countries.to_csv('ProcessedData/Countries.csv')
countries
```
